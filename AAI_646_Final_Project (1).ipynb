# Notes
* **data1-4** are portions of the raw dataset (>2.5m entries).
* **training** and **testing** are smaller, pre-cleaned portions of the raw dataset (~175k and ~82k entries respectively).
* **ground truth** is a label generator for the raw dataset, which can be used to convert data1-4 into new train-test instances.
* **summarize_dataframe** displays important values from the inputted dataframe, such as #rows, #columns, missing values, duplicated entries, distribution of attacks vs normal entries (1 = attack, 0 = normal), and distribution of attack types (DoS, Backdoor, Worms, etc.).

```python
# Load CSV files
raw_data1 = pd.read_csv('/root/.cache/kagglehub/datasets/freshersstaff/unsw-nb15-dataset/versions/1/UNSW-NB15_c/UNSW-NB15_1.csv', header=None)
raw_data2 = pd.read_csv('/root/.cache/kagglehub/datasets/freshersstaff/unsw-nb15-dataset/versions/1/UNSW-NB15_c/UNSW-NB15_2.csv', header=None)
raw_data3 = pd.read_csv('/root/.cache/kagglehub/datasets/freshersstaff/unsw-nb15-dataset/versions/1/UNSW-NB15_c/UNSW-NB15_3.csv', header=None)
raw_data4 = pd.read_csv('/root/.cache/kagglehub/datasets/freshersstaff/unsw-nb15-dataset/versions/1/UNSW-NB15_c/UNSW-NB15_4.csv', header=None)
# Note: Training and Testing datasets are mislabeled!
raw_testing = pd.read_csv('/root/.cache/kagglehub/datasets/freshersstaff/unsw-nb15-dataset/versions/1/UNSW-NB15_c/UNSW_NB15_training-set.csv')
raw_training = pd.read_csv('/root/.cache/kagglehub/datasets/freshersstaff/unsw-nb15-dataset/versions/1/UNSW-NB15_c/UNSW_NB15_testing-set.csv')
raw_gt = pd.read_csv('/root/.cache/kagglehub/datasets/freshersstaff/unsw-nb15-dataset/versions/1/UNSW-NB15_c/NUSW-NB15_GT.csv')

# Combine raw_data1-4 into a single DataFrame
raw_data_full = pd.concat([raw_data1, raw_data2, raw_data3, raw_data4], ignore_index=True)

# Display the first 5 rows of each file
print("\n--- Raw Data 1-4 Head ---")
display(raw_data_full.head())
print("\n--- Raw Training Head ---")
display(raw_training.head())
print("\n--- Raw Testing Head ---")
display(raw_testing.head())
print("\n--- Raw Ground Truth Head ---")
display(raw_gt.head())
```

```python
def summarize_dataframe(df: pd.DataFrame) -> dict:
    """Create a compact profile that is useful for first-pass exploration."""
    label_counts = df["label"].value_counts(dropna=False).to_dict() if "label" in df.columns else {}
    attack_counts = df["attack_cat"].value_counts(dropna=False).to_dict() if "attack_cat" in df.columns else {}

    return {
        "rows": int(df.shape[0]),
        "columns": int(df.shape[1]),
        "missing_values_per_column": df.isna().sum().sort_values(ascending=False).to_dict(),
        "duplicate_rows": int(df.duplicated().sum()),
        "label_distribution": label_counts,
        "attack_category_distribution": attack_counts,
    }
```

```python
# Call summarize_dataframe for all dataframes
print("\n--- Raw Data 1-4 Summary ---")
display(summarize_dataframe(raw_data_full))

print("\n--- Raw Training Summary ---")
display(summarize_dataframe(raw_training))

print("\n--- Raw Testing Summary ---")
display(summarize_dataframe(raw_testing))

print("\n--- Raw Ground Truth Summary ---")
display(summarize_dataframe(raw_gt))
```

# Preprocess Training / Testing Data

```python
# Perform ordinal and one-hot encoding on non-numeric features:
# > Ordinal encoding is applied to proto.
# > One-hot encoding is applied to service, state, and attack_cat.

# Copy data for preprocessing
encoded_training = raw_training.copy()
encoded_testing = raw_testing.copy()

# Ordinal Encoding.
# > pd.factorize() is used since there is no explicit hierarchy (uses only the indexes).
encoded_training['proto'] = pd.factorize(encoded_training['proto'])[0]
encoded_testing['proto'] = pd.factorize(encoded_testing['proto'])[0]

# One-Hot Encoding
encoded_training = pd.get_dummies(encoded_training, columns=['service', 'state', 'attack_cat'], drop_first=True)
encoded_testing = pd.get_dummies(encoded_testing, columns=['service', 'state', 'attack_cat'], drop_first=True)

# Reindex training-testing due to #columns mismatch after encoding.
columns = list(set(encoded_training.columns) | set(encoded_testing.columns))
encoded_training = encoded_training.reindex(columns=columns, fill_value=0)
encoded_testing = encoded_testing.reindex(columns=columns, fill_value=0)

print("\n--- Encoded Training Head ---")
display(encoded_training.head())

print("\n--- Encoded Testing Head ---")
display(encoded_testing.head())
```

```python
# Perform standardization using StandardScaler

# Identify numeric columns
numeric_train_columns = encoded_training.select_dtypes(include=['number']).columns.tolist()
numeric_test_columns = encoded_testing.select_dtypes(include=['number']).columns.tolist()

# Filter out 'label' column (since this is the target var)
numeric_train_columns.remove('label')
numeric_test_columns.remove('label')

# Filter out one-hot encoded columns (since these are boolean 0-1)
# > Loop through the numeric columns, determining which need to be dropped.
# > Drop columns AFTER these loops to prevent indexing issues.
remove_train_columns = []
remove_test_columns = []

for col in numeric_train_columns:
    if col.startswith('service_') or col.startswith('state_') or col.startswith('attack_cat_'):
        remove_train_columns.append(col)
for col in numeric_test_columns:
    if col.startswith('service_') or col.startswith('state_') or col.startswith('attack_cat_'):
        remove_test_columns.append(col)

for col in remove_train_columns:
    numeric_train_columns.remove(col)
for col in remove_test_columns:
    numeric_test_columns.remove(col)

# Scale train-test data
scaler = StandardScaler()
encoded_training[numeric_train_columns] = scaler.fit_transform(encoded_training[numeric_train_columns]) # Fit & transform
encoded_testing[numeric_test_columns] = scaler.transform(encoded_testing[numeric_test_columns])         # Transform only

print("\n--- Scaled Training Head ---")
display(encoded_training.head())

print("\n--- Scaled Testing Head ---")
display(encoded_testing.head())
```

# Feature Selection

```python
# Copy data for feature selection
fs_training = encoded_training.copy()
fs_testing = encoded_testing.copy()

# Drop 'id' column (since this is just the index)
fs_training = fs_training.drop('id', axis=1)
fs_testing = fs_testing.drop('id', axis=1)

print("\n--- Training Head w/o id ---")
display(fs_training.head())

print("\n--- Testing Head w/o id ---")
display(fs_testing.head())
```
